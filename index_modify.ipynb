```python
import pickle as pickle
import os
import pandas as pd
import torch
import numpy as np
import re

import sys
import numpy as np
import pandas as pd
import random
from tqdm.notebook import tqdm
import os
import re
import time
from collections import Counter

from ast import literal_eval
```

```python
df = pd.read_csv('../dataset/train/train.csv')
df.head()
```

```python
id_list = []
sentence_list = []
subject_list = []
object_list = []
label_list = []
source_list = []
id_number = 0

for idx, data in df.iterrows():
    id_list.append(str(id_number))
    sentence_list.append(data['sentence'])
    label_list.append(data['label'])
    source_list.append(data['source'])

    subject_word = literal_eval(data['subject_entity'])['word'].strip()
    subject_type = literal_eval(data['subject_entity'])['type']
    subject_start_idx = data['sentence'].find(subject_word)
    subject_end_idx = subject_start_idx + len(subject_word) - 1
    subject_entity_dict = {"word":subject_word, "start_idx":subject_start_idx, "end_idx":subject_end_idx, "type":subject_type}
    subject_list.append(str(subject_entity_dict))

    object_word = literal_eval(data['object_entity'])['word'].strip()
    object_type = literal_eval(data['object_entity'])['type']
    object_start_idx = data['sentence'].find(object_word)
    object_end_idx = object_start_idx + len(object_word) - 1
    object_entity_dict = {"word":object_word, "start_idx":object_start_idx, "end_idx":object_end_idx, "type":object_type}
    object_list.append(str(object_entity_dict))

    id_number += 1
```

```python
data = {"id":id_list, "sentence":sentence_list, "subject_entity":subject_list, "object_entity":object_list,"label":label_list,"source":source_list } 
df = pd.DataFrame(data, columns = ['id', 'sentence' , 'subject_entity', 'object_entity', 'label', 'source'])
df.to_csv("../dataset/train/train.csv", index=False, mode='a', encoding='utf-8', header=True)
```
